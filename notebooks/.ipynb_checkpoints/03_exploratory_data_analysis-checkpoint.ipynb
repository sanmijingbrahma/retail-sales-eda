# Exploratory Data Analysis

## Objective

## Import Library and set up data

```python
import pandas as pd

df = pd.read_csv("../data/processed/retail_sales_cleaned.csv")

df.head()
```

## What is the Total Revenue?

```python
df["Total Amount"].sum()
```

### Business Insight
- Dataset reveals that retail has total revenue of 456,000

## Which Product Category generates the highest revenue?

```python
df.groupby("Product Category")["Total Amount"].sum()
```

### Business Insight
- Electronics Category has highest reveue
- Electronis has reveune of 156905
- Clothing comes in seconde for the revenue


## Which Product Category has most transactions?

```python
df["Product Category"].value_counts()
```

### Business Insight
- Clothing has most transaction in dataset
- With total 351 transactions on it

## Average Customer Age

```python
df["Age"].mean()
```

### Business Insight
- The average age of customer is approximately 41.392 
- Which could mean older individual is average customer for our retail


## Gender Distribution of Customers

```python
df["Gender"].value_counts()
```

### Business insights
- Our retail dataset has higher female customer distribution
