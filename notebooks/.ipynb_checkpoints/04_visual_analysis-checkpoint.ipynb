# Visual Analysis of Dataset using Matplotlib
## Objective
- Understanding data more better using data visualization
- Using bar charts, histogram, line charts to analyze data

## Import Libraries and setup data

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/retail_sales_cleaned.csv",parse_dates=["Date"])
df.dtypes
```

## Total Revenue by Product Category

```python
category_sales = df.groupby("Product Category")["Total Amount"].sum()

plt.bar(category_sales.index,category_sales.values)
plt.title("Total Revenue by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Amount")
plt.show()
```

### Visual Insights
- Beauty products have smallest contribution to total revenue as per data compared to clothing and electronics
- Hithest revenue is by Electronics products

```python
category_transaction = df["Product Category"].value_counts()


plt.bar(category_transaction.index,category_transaction.values)
plt.title("Total Transaction by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Number of Transcation")
plt.show()
```

### Visual Insight
- Beauty product have lowest number of transaction which might have lead to low contribution to total revenue
- Additionaly here clothing has more transaction but it was not highest contributer in total revenue
- So we can say that electronics have high value products compared to clothing

## Customer Gender Distribution

```python
gender_count = df["Gender"].value_counts()

plt.bar(gender_count.index,gender_count.values)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Frequency")
plt.show()
```

### Visual Insights 
- Dataset have almost equal distribution of gender with only difference of 20 datapoints, indicating female majority customers.


## Customer Age Distribution

```python
plt.hist(df["Age"])
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

### Visual Insights
- Here in Customer age distribution is it almost normal distribuition of data

## Monthly Revenue Trend

```python
monthly_revenue = df.groupby(df["Date"].dt.to_period("M"))["Total Amount"].sum()
monthly_revenue

plt.figure(figsize=(15,5))
plt.plot(monthly_revenue.index.astype(str),monthly_revenue.values)
plt.title("Monthly Revenue Trend")
plt.xlabel("Months")
plt.ylabel("Revenue")
plt.show()
```

### Visual Insights
- Here we have seen that there was hights revenue in month 5th 
- But after that it went down in 6th,7th,8th and 9th continuously
