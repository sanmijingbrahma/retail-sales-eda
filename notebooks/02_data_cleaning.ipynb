# Clean the retail dataset
## Objective
- Convert Date column to datetime
- Check every numeric column has correct numeric datatype
- Check for impossible values
    - Age < 0
    - Quantity < 0
    - Price per Unit < 0
    - Total Amount < 0
- Save the cleaned dataset

## Import Library and set up data

```python
import pandas as pd

df = pd.read_csv("../data/raw/retail_sales_dataset.csv")

df.head()
```

## Convert Date to Datetime

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.dtypes
```

- Here we have successfully converted "Date" to datetime datatype

## Check For Impossible Values

```python
(df["Age"] <= 0).sum()
```

```python
(df["Quantity"] <= 0).sum()
```

```python
(df["Price per Unit"] <= 0).sum()
```

```python
(df["Total Amount"] <= 0).sum()
```

- Here hwe have found out that all numerical values do not have any impossible values

## Saving the cleaned dataset

```python
df.to_csv("../data/processed/retail_sales_cleaned.csv",index=False)
```

## Conclusion
- Date datatype has been converted to datetime
- We have checked for any impossible values for numerical data
- As we have already checked for missing and duplicate value so we have not done it
- Then we have saved the cleaned dataset
