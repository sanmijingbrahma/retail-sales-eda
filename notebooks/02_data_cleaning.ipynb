# Clean the retail dataset
## Objective
- Convert Date column to datetime
- Check every numeric column has correct numeric datatype
- Check for impossible values
    - Age < 0
    - Quantity < 0
    - Price per Unit < 0
    - Total Amount < 0
- Save the cleaned dataset

## Import Library and set up data

```python
import pandas as pd

df = pd.read_csv("../data/raw/retail_sales_dataset.csv")

df.head()
```

## Convert Date to Datetime

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.dtypes
```

- Here we have successfully converted "Date" to datetime datatype

## Check For Missing Values

```python
df.isna().sum()
```

## Check fo Duplicate Values

```python
df.duplicated().sum()
```

## Check For Impossible Values

```python
cols_to_check = ["Age","Quantity","Price per Unit","Total Amount"]

summary = {
    "Columns":cols_to_check,
    "Impossible(<0)": [(df[cols]<=0).sum() for cols in cols_to_check]
}

impossible_summary = pd.DataFrame(summary)

impossible_summary
```

- Here hwe have found out that all numerical values do not have any impossible values

## Saving the cleaned dataset

```python
df.to_csv("../data/processed/retail_sales_cleaned.csv",index=False)
df.dtypes
```

## Conclusion
- So we have found out that there is no missing and duplicate values
- Also there were not impossible value for the numerical columns
- As there was "Date" in type object we have converted it into "datetime" datatype
- Andd also data has been cleaned and saved
